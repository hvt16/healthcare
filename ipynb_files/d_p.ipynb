```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
train_data = pd.read_csv("../datasets/archive/Training.csv")
test_data = pd.read_csv("../datasets/archive/Testing.csv")
train_data.head(20)
```

```python
# features
x_train = train_data.drop(columns=["prognosis"]).dropna(axis=1, how="any", thresh=None, subset=None, inplace=False)

# labels
y_train = train_data["prognosis"]
```

```python
symptomes = list(x_train.columns)
print('total symptomes :',len(symptomes))
# symptomes.sort()
print(symptomes)
```

```python
diseases = list(set(y_train))
print('total diseases :', len(diseases))
diseases.sort()
print(diseases)
```

```python
x_test = test_data.drop(columns=["prognosis"])
y_test = test_data["prognosis"]
```

```python
# decision tree classifier
model_dtc = DecisionTreeClassifier()

# random forest classifier
model_rfc = RandomForestClassifier()
```

```python
model_dtc.fit(x_train, y_train)
```

```python
prediction = model_dtc.predict(x_test)
```

```python
prediction
```

```python
accuracy = accuracy_score(prediction, y_test)
print(accuracy)
```

```python
# symptomes list on which the disease is to be predicted
def df_sym(syms):
    di = {}
    for e in symptomes:
        di[e] = 0
    for e in syms:
        if e in di:
            di[e] = 1
    test = pd.DataFrame(di, index=[0])
    return test

my_sym = ['excessive_hunger', 'polyuria', 'irregular_sugar_level', 'weight_loss']

test = df_sym(list(('itching','skin_rash','nodal_skin_eruptions')))
```
